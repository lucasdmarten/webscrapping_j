```python
import time
import requests
from bs4 import BeautifulSoup
import pprint
import pandas as pd
import datetime
import csv
import matplotlib.pyplot as plt
import numpy as np

import metpy.calc as mpcalc
from metpy.plots import  SkewT
from metpy.units import units
```

```python
# url = driver.current_url
r = requests.get(("http://weather.uwyo.edu/cgi-bin/sounding?region=samer&TYPE=TEXT%3ALIST&YEAR=2018&MONTH=04&FROM=2300&TO=2500&STNM=83937"))
soup = BeautifulSoup(r.content, 'html.parser')
pre=soup.find_all('pre')
h2=soup.find_all('h2')
```

```python
n=0
i=0
while n<10:
    # get data obj
    date=h2[i]
    
    l_info = date.text.split(' ')
    day, hour = int(l_info[-3]),int(l_info[-4][:-1])
    
    with open('sounding.c1_{}_{}Z.csv'.format(day,hour), 'w', newline='') as csvfile:
        write = csv.writer(csvfile, delimiter=' ',
                                quotechar='|', quoting=csv.QUOTE_MINIMAL)
        write.writerow(pre[n])
    
    
    i+=1
    n+=2
```

```python
names=['pressure','height','temp','dewpoint','humidity_r','moisture','dir','spdknt','THTA','THTE','THTV']
df=pd.read_csv('./sounding.c1_24_0Z.csv',skiprows=5,delimiter='\s+',names=names)
```

```python
p, td, t = df.pressure.interpolate().astype('float').values * units.hPa, df.dewpoint.interpolate().values * units.degC, df.temp.interpolate().values * units.degC
```

```python
wind_speed = df['spdknt'].interpolate().astype('float').values * units.knots
wind_dir = df['dir'].interpolate().astype('float').values * units.degrees
u, v = mpcalc.wind_components(wind_speed, wind_dir)
```

```python
plt.rcParams['figure.figsize'] = (9, 9)
fig = plt.figure()

skew = SkewT(fig, rotation=45)

skew.plot(p, t, 'r')
skew.plot(p, td, 'g')
skew.plot_barbs(p, u, v)
skew.ax.set_ylim(1000, 100)

skew.plot_dry_adiabats()
skew.plot_moist_adiabats()
skew.plot_mixing_lines()

skew.ax.set_xlim(-50, 60)

fig = plt.gcf()
```
